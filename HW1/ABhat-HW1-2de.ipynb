```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
theta = 0.6
c = 2.5
estimateSize = 1000
```

```python
def calculateProbability(n):
    limit = 0
    for _ in range(estimateSize):
        p = np.random.choice([0, 2, 4], size=n, p=[(1 - theta) / 2, theta, (1 - theta) / 2])
        p_mean = np.mean(p)
        if p_mean >= c:
            limit += 1
    return limit / estimateSize
```

```python
n_values = np.array([i for i in range(10,1001,10)])

emp_estimates = []
for n in n_values:
    emp_estimates.append(calculateProbability(n))
    
markov = 2 / (c * np.ones_like(n_values))
chebyshev = (4 - 4 * theta) / (n_values * ((c - 2) ** 2))
hoeffding = np.exp(-n_values * ((c - 2) ** 2) / 8)
```

<h4> Upper Bound Estimate for P[Mₙ(X) ≥ 2.5] </h4>
c = 2.5, 
θ = 0.6

```python
# c = 2.5
plt.plot(n_values, markov, label='Markov')
plt.plot(n_values, chebyshev, label='Chebyshev')
plt.plot(n_values, hoeffding, label='Hoeffding')
plt.plot(n_values, emp_estimates, label='Empirical Estimate')

plt.title(f'Upper Bound Estimate for P[Mₙ(X) ≥ {c}]')
plt.xlabel('n')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
```

```python
c = 3.0
n_values = np.array([i for i in range(10,1001,10)])

emp_estimates = []
for n in n_values:
    emp_estimates.append(calculateProbability(n))
    
markov = 2 / (c * np.ones_like(n_values))
chebyshev = (4 - 4 * theta) / (n_values * ((c - 2) ** 2))
hoeffding = np.exp(-n_values * ((c - 2) ** 2) / 8)
```
