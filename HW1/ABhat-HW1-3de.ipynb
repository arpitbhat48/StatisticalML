```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mu = 5.0  
sigma = 1
c = mu + sigma 
```

```python
n_values = np.array([i for i in range(10,1001,10)])

chebyshev = (sigma**2) / ((c - mu)**2)
emp_estimates = np.zeros(len(n_values))
```

```python
def calculate_probability(sample, c):
    return np.mean(sample >= c)
```

```python
for i, n in enumerate(n_values):
    sample = np.random.normal(mu, sigma, size=n)
    emp_estimates[i] = calculate_probability(sample, c)
```

<h4> Upper Bound Estimate for P[Mₙ(X) ≥ 6.0] </h4>
c = 6.0, 

```python
plt.plot(n_values, chebyshev * np.ones_like(n_values), label='Chebyshev Bound')
plt.plot(n_values, emp_estimates, label='Empirical Estimate')
plt.plot(n_values, np.mean(emp_estimates)*np.ones_like(n_values), label='Mean Estimate')

plt.xlabel('n')
plt.ylabel('Probability')
plt.title(f'Bounds and Empirical Estimate for P[Mₙ(X) ≥ {c}]')
plt.legend()
plt.grid(True)
```

<p>
Chebyshev Bound: It does not depend on size n. The bound is constant and depicts the worst case bound. This bound depicts the weak law of large numbers.
    <br>
Empirical Estimate: It depends on size n, and the estimate becomes less noisy with increasing n.
</p>
