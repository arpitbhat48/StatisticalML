<h3>9.b. 3D Hyperplane </h3>

w = [1,1,1]<sup>T</sup> 
<br>
    b = 0
<br>
    w<sub>1</sub>x + w<sub>2</sub>y + w<sub>3</sub>z + b = 0
<br>
    x + y + z = 0


```python
x = np.arange(-10,10,0.5)
y = np.arange(-10,10,0.5)
X, Y = np.meshgrid(x,y)
Z = X + Y
```

```python
ax = plt.axes(projection="3d")
ax.plot_surface(X,Y,Z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.title("3D Hyper Plane")
```
